```python
import tkinter as tk
from tkinter import ttk
win=tk.Tk()
win.title('파이썬')
win.geometry('650x150')
win.focus_force()
r0c0=ttk.Label(win,text='글자써보자')
r0c0.grid(row=0,column=0)

r1c0=ttk.Button(win,text='클릭')
r1c0.grid(row=1,column=0)
num=0
def countUp():
    global num
    num+=1
    r1c0['text']=num
r1c0.config(command=countUp)

number=tk.StringVar()
number2=tk.StringVar()
r2c0=ttk.Entry(win,textvariable=number,width=10)
r2c0.grid(row=2,column=0)
r2c1=ttk.Label(win,text='+')

r2c2=ttk.Entry(win,textvariable=number2,width=10)
r2c2.grid(row=2,column=2)
r2c3=ttk.Button(win,text='=',width=3)

r3c0=ttk.Button(win,text='1',width=3)
r3c0.grid(row=3,column=0)
r3c1=ttk.Button(win,text='2',width=3)
r3c1.grid(row=3,column=1)
r3c2=ttk.Button(win,text='3',width=3)
r3c2.grid(row=3,column=2)
r3c3=ttk.Button(win,text='4',width=3)
r3c3.grid(row=3,column=3)
r3c4=ttk.Button(win,text='5',width=3)
r3c4.grid(row=3,column=4)

r3c5=ttk.Button(win,text='1번 입력',width=7)
r3c5.grid(row=3,column=5)
r3c6=ttk.Button(win,text='2번 입력',width=7)
r3c6.grid(row=3,column=6)
r3c7=ttk.Label(win,text='1')

r4c0=ttk.Button(win,text='6',width=3)
r4c0.grid(row=4,column=0)
r4c1=ttk.Button(win,text='7',width=3)
r4c1.grid(row=4,column=1)
r4c2=ttk.Button(win,text='8',width=3)
r4c2.grid(row=4,column=2)
r4c3=ttk.Button(win,text='9',width=3)
r4c3.grid(row=4,column=3)
r4c4=ttk.Button(win,text='0',width=3)
r4c4.grid(row=4,column=4)

r5c0=ttk.Button(win,text='+',width=3)
r5c0.grid(row=5,column=0)
r5c1=ttk.Button(win,text='-',width=3)
r5c1.grid(row=5,column=1)
r5c2=ttk.Button(win,text='*',width=3)
r5c2.grid(row=5,column=2)
r5c3=ttk.Button(win,text='/',width=3)
r5c3.grid(row=5,column=3)
r5c4=ttk.Button(win,text='C',width=3)
r5c4.grid(row=5,column=4)

def sum(a,b):
    c=a+b
    r2c4=ttk.Label(win,text=str(c))
    r2c4.grid(row=2,column=4)
def minus(a,b):
    c=a-b
    r2c4=ttk.Label(win,text=str(c))
    r2c4.grid(row=2,column=4)
def time(a,b):
    c=a*b
    r2c4=ttk.Label(win,text=str(c))
    r2c4.grid(row=2,column=4)
def division(a,b):
    c=a/b
    r2c4=ttk.Label(win,text=str(c))
    r2c4.grid(row=2,column=4)
def action(a,b,c):
    r2c4=ttk.Label(win,text='          ',width=10)
    r2c4.grid(row=2,column=4)
    if(c=='+'):
        return sum(a,b)
    elif(c=='-'):
        return minus(a,b)
    elif(c=='*'):
        return time(a,b)
    elif(c=='/'):
        return division(a,b)

def sumclick():
    r2c1['text']='+'
    r2c1.grid(row=2,column=1)
r5c0.config(command=sumclick)
def minusclick():
    r2c1['text']='-'
    r2c1.grid(row=2,column=1)
r5c1.config(command=minusclick)
def timeclick():
    r2c1['text']='*'
    r2c1.grid(row=2,column=1)
r5c2.config(command=timeclick)
def divisionclick():
    r2c1['text']='/'
    r2c1.grid(row=2,column=1)
r5c3.config(command=divisionclick)
r2c3.config(command=lambda:action(int(number.get()),int(number2.get()),r2c1['text']))

def oneaddclick():
    r3c7['text']='1'
    r3c7.grid(row=3,column=7)
r3c5.config(command=oneaddclick)
def twoaddclick():
    r3c7['text']='2'
    r3c7.grid(row=3,column=7)
r3c6.config(command=twoaddclick)


def oneclick(a):
    if(a=='1'):
        r2c0.insert(tk.END,'1')
        r2c0.grid(row=2,column=0)
    elif(a=='2'):
        r2c2.insert(tk.END,'1')
        r2c2.grid(row=2,column=2)
r3c0.config(command=lambda:oneclick(r3c7['text']))
def twoclick(a):
    if(a=='1'):
        r2c0.insert(tk.END,'2')
        r2c0.grid(row=2,column=0)
    elif(a=='2'):
        r2c2.insert(tk.END,'2')
        r2c2.grid(row=2,column=2)
r3c1.config(command=lambda:twoclick(r3c7['text']))
def threeclick(a):
    if(a=='1'):
        r2c0.insert(tk.END,'3')
        r2c0.grid(row=2,column=0)
    elif(a=='2'):
        r2c2.insert(tk.END,'3')
        r2c2.grid(row=2,column=2)
r3c2.config(command=lambda:threeclick(r3c7['text']))
def fourclick(a):
    if(a=='1'):
        r2c0.insert(tk.END,'4')
        r2c0.grid(row=2,column=0)
    elif(a=='2'):
        r2c2.insert(tk.END,'4')
        r2c2.grid(row=2,column=2)
r3c3.config(command=lambda:fourclick(r3c7['text']))
def fiveclick(a):
    if(a=='1'):
        r2c0.insert(tk.END,'5')
        r2c0.grid(row=2,column=0)
    elif(a=='2'):
        r2c2.insert(tk.END,'5')
        r2c2.grid(row=2,column=2)
r3c4.config(command=lambda:fiveclick(r3c7['text']))
def sixclick(a):
    if(a=='1'):
        r2c0.insert(tk.END,'6')
        r2c0.grid(row=2,column=0)
    elif(a=='2'):
        r2c2.insert(tk.END,'6')
        r2c2.grid(row=2,column=2)
r4c0.config(command=lambda:sixclick(r3c7['text']))
def sevenclick(a):
    if(a=='1'):
        r2c0.insert(tk.END,'7')
        r2c0.grid(row=2,column=0)
    elif(a=='2'):
        r2c2.insert(tk.END,'7')
        r2c2.grid(row=2,column=2)
r4c1.config(command=lambda:sevenclick(r3c7['text']))
def eightclick(a):
    if(a=='1'):
        r2c0.insert(tk.END,'8')
        r2c0.grid(row=2,column=0)
    elif(a=='2'):
        r2c2.insert(tk.END,'8')
        r2c2.grid(row=2,column=2)
r4c2.config(command=lambda:eightclick(r3c7['text']))
def nineclick(a):
    if(a=='1'):
        r2c0.insert(tk.END,'9')
        r2c0.grid(row=2,column=0)
    elif(a=='2'):
        r2c2.insert(tk.END,'9')
        r2c2.grid(row=2,column=2)
r4c3.config(command=lambda:nineclick(r3c7['text']))
def zeroclick(a):
    if(a=='1'):
        r2c0.insert(tk.END,'0')
        r2c0.grid(row=2,column=0)
    elif(a=='2'):
        r2c2.insert(tk.END,'0')
        r2c2.grid(row=2,column=2)
r4c4.config(command=lambda:zeroclick(r3c7['text']))
def clearclick():
    r2c0.delete(0,tk.END)
    r2c2.delete(0,tk.END)
r5c4.config(command=clearclick)


r2c3.grid(row=2,column=3)

win.mainloop()
```
